```python
def get_antonyms(word, tag):
    synonyms = []
    antonyms = []

    for syn in wordnet.synsets(word, tag):
        for l in syn.lemmas():
            synonyms.append(l.name())
            if l.antonyms():
                antonyms.append(l.antonyms()[0].name())
    return antonyms
```

```python
from nltk.corpus import wordnet

def get_wordnet_pos(treebank_tag):

    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return ''
```

```python
stories = pd.read_csv('/Users/shivendra/Projects/roc_stories/100KStories.csv')
```

```python
def convert_complete_sentence_into_opposite(current):
    sentence = ""    
    text = word_tokenize(current)
    pos_tags = nltk.pos_tag(text)
    
    for (word,tag) in pos_tags:
        antonym_list = get_antonyms(word, get_wordnet_pos(tag))
        if len(antonym_list)>0:
            sentence = sentence + (antonym_list[0]) + " "
        else:
            sentence = sentence + word + " "
    sentence = sentence.replace("_", " ")
    return sentence.strip()
```

```python
stories['all_opposites'] = np.vectorize(convert_complete_sentence_into_opposite)(stories['sentence5'])
```
